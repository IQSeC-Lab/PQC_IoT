```python
import pandas as pd
import glob

# Get all CSV files in the current directory
csv_files = glob.glob("server/*.csv")

for file in csv_files:
    print(f"Processing {file}...")
    try:
        df = pd.read_csv(file)
        
        # Only update if 'timestamp' column exists
        if "timestamp" in df.columns:
            df["timestamp"] = pd.to_datetime(df["timestamp"]).dt.strftime("%H:%M:%S")
            df.to_csv(file, index=False)
            # print(df.head())
            
        else:
            print(f"  Skipped: 'timestamp' column not found in {file}")
    except Exception as e:
        print(f"  Error processing {file}: {e}")

print("All files processed.")
```

```python
import pandas as pd
def clean_csv(filepath):
    """
    Loads a CSV into a pandas DataFrame and removes specific columns.
    
    Parameters:
        filepath (str): The path to the CSV file.
    
    Returns:
        pd.DataFrame: Cleaned DataFrame with specified columns removed.
    """
    cols_to_remove = ['idle_mem_kb']
    
    try:
        df = pd.read_csv(filepath)
        df.drop(columns=[col for col in cols_to_remove if col in df.columns], inplace=True)
        return df
    except Exception as e:
        print(f"Error loading or cleaning CSV: {e}")
        return None
```

```python

# Get all CSV files in the current directory
csv_files = glob.glob("server/*.csv")
for files in csv_files:
    print(f"Cleaning {files}...")
    cleaned_df = clean_csv(files)
    if cleaned_df is not None:
        cleaned_df.to_csv(files, index=False)
```
